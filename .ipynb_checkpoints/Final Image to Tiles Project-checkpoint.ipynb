```python
#convert all values in the matplotlib data object from RGB to HSBK and create new tcs.
target_tcs = []
for row in data:
    temp_row = []
    for pixel in row:
        temp_row.append(RGBtoHSBK(pixel))
    target_tcs.append(temp_row)
        
```

```python
#split 2D target_tcs into a list of 4 x 2D (8x40) Tilechains
tcsplit = tiles.split_tilechains(target_tcs)
```

```python
#combined function

# resize image and force a new shape and save to disk
def display_image(image_to_display):
    # load the image
    my_image = Image.open(image_to_display)
    # report the size of the image
    print(my_image.size)
    # resize image and ignore original aspect ratio
    img_resized = my_image.resize((40,32))
    img_resized.save('./images/resized_image.jpg')
    data = image.imread('./images/resized_image.jpg')
    target_tcs = []
    for row in data:
        temp_row = []
        for pixel in row:
            temp_row.append(RGBtoHSBK(pixel))
        target_tcs.append(temp_row)
    tcsplit = tiles.split_tilechains(target_tcs)
    T1.set_tilechain_colors(tiles.split_combined_matrix(tcsplit[0]))
    T2.set_tilechain_colors(tiles.split_combined_matrix(tcsplit[1]))
    T3.set_tilechain_colors(tiles.split_combined_matrix(tcsplit[2]))
    T4.set_tilechain_colors(tiles.split_combined_matrix(tcsplit[3]))
```
