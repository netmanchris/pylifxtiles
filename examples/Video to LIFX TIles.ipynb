```python
# Opens the Video file
cap= cv2.VideoCapture(0)
i=1
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == False:
        break
    if i%10 == 0:
        image_name = './rtsp/img'+str(i)+'.jpg'
        cv2.imwrite(image_name,frame)
        actions.display_jpg_image((image_name), (40, 40), tc_list)     
    i+=1
 
cap.release()
cv2.destroyAllWindows()
```
